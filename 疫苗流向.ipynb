```python
%matplotlib inline
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#导入数据

prov_names =  [] 
pd_alldata = pd.DataFrame()
 
mypath = "./data/"
f_names = os.listdir(mypath)

for f_names_i in f_names:
    prov_names.append(f_names_i.split(".")[0])
    f=open(mypath+ f_names_i,'rb')
    t_prov = pd.read_csv(f) 
    pd_alldata= pd_alldata.append( t_prov )
pd_alldata.head()
```

```python

pd_alldata=pd_alldata[["create_company","name","price","prov","report_company","src","year"]]
pd_alldata.head(5)
```

```python
pd_alldata["size"]=1
# pro_name_gp = pd_alldata.groupby(["prov"],sort=True).size()
# # pro_name_gp.plot(kind="box")
# pro_name_gp
pd_alldata
```

```python
def fcqcs(x):
    x = str(x)
    if  '长春长生' in x :
        return 1
    else:
        return 0 
pd_alldata['size_cqcs']= pd_alldata['create_company']
pd_alldata['size_cqcs'] = pd_alldata['size_cqcs'].apply(fcqcs)
prov_sum = pd_alldata[['prov','size','size_cqcs']].groupby(["prov"]).sum()
# prov_sum = prov_sum.reset_index()
prov_sum
```
