```python
# %matpltlib inline
import os
import pandas as pd
import numpy as np
#导入数据

prov_names =  [] 
pd_alldata = pd.DataFrame()
 
mypath = "./data/"
f_names = os.listdir(mypath)

for f_names_i in f_names:
    prov_names.append(f_names_i.split(".")[0])
    f=open(mypath+ f_names_i,'rb')
    t_prov = pd.read_csv(f) 
    pd_alldata= pd_alldata.append( t_prov )
print(pd_alldata)
```
