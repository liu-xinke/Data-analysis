position size growth color

```python
import numpy as np
import matplotlib.animation as ma
import matplotlib.pyplot as plt
%matplotlib inline
n_bubbles=100
bubbles = np.zeros(n_bubbles,dtype=[
    ('position',float,2),
    ('size',float),
    ('growth',float),
    ('color',float,4)
])
bubbles['position']=np.random.uniform(0,1,(n_bubbles,2))
bubbles['size']=np.random.uniform(50,750,n_bubbles)
bubbles['growth']=np.random.uniform(30,150,n_bubbles)
bubbles['color']=np.random.uniform(0,1,(n_bubbles,4))
sc=plt.scatter(bubbles['position'][:,0],bubbles['position'][:,1],s=bubbles['size'],c=bubbles['color'])
def update(number):
    bubbles['size']+=bubbles['growth']
    burst=number% n_bubbles
    bubbles['position'][burst]=np.random.uniform(0,1,2)
    bubbles["size"][burst]=0
    bubbles["growth"][burst]=np.random.uniform(30,150)
    sc.set_size(bubbles['size'])
anim = ma.FuncAnimation(plt.gcf(),update,interval=10)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
%matplotlib inline

# 定义figure
fig, ax = plt.subplots()

# 定义数据
x = np.arange(0, 2 * np.pi, 0.01)
# line, 表示只取返回值中的第一个元素
line, = ax.plot(x, np.sin(x))

# 定义动画的更新
def update(i):
    line.set_ydata(np.sin(x + i/10))
    return line,

# 定义动画的初始值
def init():
    line.set_ydata(np.sin(x))
    return line,

# 创建动画
ani = animation.FuncAnimation(fig = fig, func = update, init_func = init, interval = 10, blit = False, frames = 200)

# 展示动画
plt.show()

# 动画保存
ani.save('sin.gif', writer = 'imagemagick', fps = 30, dpi = 100)
```

### 布林带

- 中轨：移动平均线 --趋势  
- 上轨：中轨+标准差
- 下轨：中轨-标准差

```python
def dmy2ymd(dmy):
    dmy=str(dmy,encoding='utf-8')
    date=dt.datetime.strptime(dmy,'%d-%m-%Y').date()
    ymd=date.strftime('%Y-%m-%d')
    return ymd
dates,closing_prices=np.loadtxt("./data/aapl.csv",delimiter=",",usecols=(1,6),
                unpack=True,dtype=np.dtype("M8[D],f8"),converters={1:dmy2ymd})

n=5
medios=np.convolve(closing_prices,np.ones(n)/n,'vaild')

stds=np.zeros(medios.size)
for i in range(medios.size):
    stds[i]=closing_prices[i:i+n].std()
stds*=2
lowers=medios-stds
uppers=medios+stds

plt.figure("Medios",facecolor='gray',figsize=(20,8))
plt.title("Medios")
plt.xlabel("Date",fontsize=14)
plt.ylabel("Closing_prices",fontsize=14)
ax=plt.gca()
ax.xaxis.set_major_locator(md.WeekdayLocator(byweekday=md.MO))
ax.xaxis.set_minor_locator(md.DayLocator())
ax.xaxis.set_major_formatter(md.DateFormatter('%Y-%m-%d'))
plt.tick_params(labelsize=10)
plt.plot(dates,closing_prices,c="orange",label="closing_prices")
plt.plot(dates[n-1:],medios,c="blue",label="Medios")
plt.plot(dates[n-1:],lowers,c="red",label="lowers")
plt.plot(dates[n-1:],uppers,c="green",label="uppers")
plt.grid(linestyle=':')
plt.legend()
# plt.gcf().autofmt_xdate()
plt.show()
```

### 线性模型

- a b c d e f g
- aA+bB+cC=d
- bA+cB+dC=e   [A B C]  g=dA+eB+fC
- cA+dB+eC=f

```python
import pandas as pd
def dmy2ymd(dmy):
    dmy=str(dmy,encoding='utf-8')
    date=dt.datetime.strptime(dmy,'%d-%m-%Y').date()
    ymd=date.strftime('%Y-%m-%d')
    return ymd
dates,closing_prices=np.loadtxt("./data/aapl.csv",delimiter=",",usecols=(1,6),
                unpack=True,dtype=np.dtype("M8[D],f8"),converters={1:dmy2ymd})

n=5
pred_prices=np.zeros(closing_prices.size-n*2+1)
for i in range(pred_prices.size):
    a=np.zeros((n,n))
    for j in range(n):
        
        
        a[j,]=closing_prices[i+j:i+j+n]
    b=closing_prices[i+n:i+n*2]
    x=np.linalg.lstsq(a,b)[0]
    pred_prices[i]=b.dot(x)
plt.figure("Line_model",facecolor='gray',figsize=(20,8))
plt.title("Line_model",fontsize=20)
plt.xlabel("Date",fontsize=14)
plt.ylabel("Closing_prices",fontsize=14)
ax=plt.gca()
ax.xaxis.set_major_locator(md.WeekdayLocator(byweekday=md.MO))
ax.xaxis.set_minor_locator(md.DayLocator())
ax.xaxis.set_major_formatter(md.DateFormatter('%Y-%m-%d'))
plt.tick_params(labelsize=10)
dates=dates.astype(md.datetime.datetime)
plt.plot(dates,closing_prices,"go-",label="closing_prices")
dates=np.append(dates,dates[-1]+pd.tseries.offsets.BDay())
plt.plot(dates[n*2:],pred_prices,'ro--',label="line_model")
plt.legend()
```

#### 线性拟合

```python
import pandas as pd
def dmy2ymd(dmy):
    dmy=str(dmy,encoding='utf-8')
    date=dt.datetime.strptime(dmy,'%d-%m-%Y').date()
    ymd=date.strftime('%Y-%m-%d')
    return ymd
dates,closing_prices=np.loadtxt("./data/aapl.csv",delimiter=",",usecols=(1,6),
                unpack=True,dtype=np.dtype("M8[D],f8"),converters={1:dmy2ymd})

n=2
pred_prices=np.zeros(closing_prices.size-n*2+1)
for i in range(pred_prices.size):
    a=np.zeros((n,n))
    for j in range(n):
        a[j][0]=dates[i]
        a[j][1]=1
    b=closing_prices[i]
    x=np.linalg.lstsq(a,b)[0]
    pred_prices[i]=b.dot(x)
plt.figure("Line_model",facecolor='gray',figsize=(20,8))
plt.title("Line_model",fontsize=20)
plt.xlabel("Date",fontsize=14)
plt.ylabel("Closing_prices",fontsize=14)
ax=plt.gca()
ax.xaxis.set_major_locator(md.WeekdayLocator(byweekday=md.MO))
ax.xaxis.set_minor_locator(md.DayLocator())
ax.xaxis.set_major_formatter(md.DateFormatter('%Y-%m-%d'))
plt.tick_params(labelsize=10)
dates=dates.astype(md.datetime.datetime)
plt.plot(dates,closing_prices,"go-",label="closing_prices")
dates=np.append(dates,dates[-1]+pd.tseries.offsets.BDay())
plt.plot(dates[n*2:],pred_prices,'ro--',label="line_model")
plt.legend()
```

# 机器学习

**回归问题**：输出数据是无限可能的连续值  
**分类问题**：输出数据是有限的几个离散值  

**无监督学习**：在输出数据未知的前提下，利用模型本身发现输入数据内部特征，将其划分为不同的族群 **聚类问题**

**半监督学习**：利用相对较少的已知集训练模型，使其获得基本的预测能力，当模型遇到未知输出的新数  
            据时，可以根据其与已知集的相似性。预测其输出

```python
import numpy as np
import matplotlib.pyplot as mp
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline
train_x = np.array([0.5, 0.6, 0.8, 1.1, 1.4])
train_y = np.array([5.0, 5.5, 6.0, 6.8, 7.0])
n_epoches = 1000
lrate = 0.01
epoches, losses = [], []
w0, w1 = [1], [1]
for epoch in range(1, n_epoches + 1):
    epoches.append(epoch)
    losses.append(((train_y - (
        w0[-1] + w1[-1] * train_x)) ** 2).sum() / 2)
#     print('{:4} w0={:.8f}, w1={:.8f}, loss={:.8f}'.format(
#         epoches[-1], w0[-1], w1[-1], losses[-1]))
    d0 = -(train_y - (
        w0[-1] + w1[-1] * train_x)).sum()
    d1 = -((train_y - (
        w0[-1] + w1[-1] * train_x)) * train_x).sum()
    w0.append(w0[-1] - lrate * d0)
    w1.append(w1[-1] - lrate * d1)
w0 = np.array(w0[:-1])
w1 = np.array(w1[:-1])
sorted_indices = train_x.argsort()
test_x = train_x[sorted_indices]
test_y = train_y[sorted_indices]
pred_test_y = w0[-1] + w1[-1] * test_x


grid_w0,grid_w1=np.meshgrid(np.linspace(0,9,500),np.linspace(0,3.5,500))

flat_w0,flat_w1 = grid_w0.ravel(),grid_w1.ravel()

flat_loss=((flat_w0+np.outer(train_x,flat_w1)-train_y.reshape(-1,1))**2).sum(axis=0)/2
grid_loss=flat_loss.reshape(grid_w0.shape)
mp.figure('Linear Regression', facecolor='lightgray')
mp.title('Linear Regression', fontsize=20)
mp.xlabel('x', fontsize=14)
mp.ylabel('y', fontsize=14)
mp.tick_params(labelsize=10)
mp.grid(linestyle=':')
mp.scatter(train_x, train_y, marker='s', c='dodgerblue',
           alpha=0.5, s=80, label='Training')
mp.scatter(test_x, test_y, marker='D', c='orangered',
           alpha=0.5, s=60, label='Testing')
mp.scatter(test_x, pred_test_y, c='orangered',
           alpha=0.5, s=80, label='Predicted')
for x, y, pred_y in zip(test_x, test_y, pred_test_y):
    mp.plot([x, x], [y, pred_y], c='orangered',
            alpha=0.5, linewidth=1)
mp.plot(test_x, pred_test_y, '--', c='limegreen',
        label='Regression', linewidth=1)
mp.legend()
mp.figure("loss funcyion")
ax=mp.gca(projection="3d")
ax.plot_surface(grid_w0,grid_w1,grid_loss,rstride=10,cmap="jet")
ax.plot(w0,w1,losses,"ro-")
mp.figure("Batch Gadient Desent",figsize=(15,8))
mp.title("Batch Gadient Desent")
mp.contourf(grid_w0,grid_w1,grid_loss,1000,cmap="jet")
cntr=mp.contour(grid_w0,grid_w1,grid_loss,10,colors="black",linewidths=0.5)
mp.clabel(cntr,inline_space=0.1,fmt="%.2f")
mp.plot(w0,w1,"ro-")
mp.show()
```


## 决策树回归和分类

- 相似的输入必会产生相似的输出
- 信息熵   -p1log2(p1)-p2log2(p2)
- 信息增益
- 完全决策树：使用所有特征作为子表划分的依据，树状结构复杂，构建和预测速度慢  
- 非完全决策树：根据信息增益最大的原则，优先选择部分特征划分子表，在输入相似的条件下，预测相似的输

集合算法：通过不同的方式构建出多棵决策树模型，分别作出预测，将他们给出的预测值通过平均或投票的方式综合考虑  
- 自助聚合：从样本空间中，以放回抽样的方式随机挑选部分样本构建决策树，共构建B棵树分别对未知样本进行预测，得出B个预测结果，平均或投票产生最后结果
- 随机森林：在自助聚合算法的基础上，每次抽样不仅随机选择样本，也随机选择特征，来构建B棵决策树，泛华不同特征对预测结果的影响  
- 正向激励：开始为每个样本分配初始权重，构建决策树，对训练集中的样本进行预测，针对预测错误的样本，增加其权重，再构建决策树，重复以上的过程，共得到B棵决策树，由这B棵树分别对未知样本进行预测，给出预测结果，再进行平均或投票

特征重要性：决策树模型在构建树状结构时，优先选择对输出结果影响最大，可得到特征重要性序列

## 聚类

K 均值   
- 给聚类中心分配点，计算所有的训练样本与每个聚类中心的距离，把每个训练样本分配到与之距离最近的中心坐在的类别中，计算每个聚类的几何中心，将聚类中心移动到几何中心上,重复以上过程，直至中心  
- 聚类数必须事先已知，通过基于聚类指标的超参数寻优解决
- 聚类中心的起始位置会影响最后的聚类效果 初始聚类中心足够分散 kmeans ++

```python
import sklearn.cluster as sc
import sklearn.metrics as sm
import matplotlib.pyplot as mp
import numpy as np
%matplotlib inline

x= []
with open("./data/multiple3.txt",'r') as f:
    for line in f.readlines():
        data=[float(substr) for substr in line.split(",")]
        x.append(data)      
x=np.array(x)
model = sc.KMeans(init="k-means++",n_clusters=4)
model.fit(x)
centers=model.cluster_centers_


l,r,h = x[:,0].min()-1,x[:,0].max()+1,0.005

b,t,v = x[:,1].min()-1,x[:,1].max()+1,0.005

grid_x=np.meshgrid(np.arange(l,r,h),np.arange(b,t,v))

flat_x=np.c_[grid_x[0].ravel(),grid_x[1].ravel()]
flat_y=model.predict(flat_x)
grid_y=flat_y.reshape(grid_x[0].shape)

pred_y=model.predict(x)

mp.pcolormesh(grid_x[0],grid_x[1],grid_y,cmap='gray')
mp.scatter(x[:,0],x[:,1],c=pred_y,cmap="brg",s=60)
mp.scatter(centers[:,0],centers[:,1],marker="+",s=200,c='red')
mp.show()
```

### 均值漂移算法

- 将待划分聚类的样本空间视作服从某种概率分布规律的随机数据，通过计算随机数据的直方图，寻找与之最为匹配的概率密度函数，该函数的峰值点就是样本中心  
- 适用于概率特征明显的聚类样本，无需事先给出聚类数，模型可以自动识别聚类个数

```python
import sklearn.cluster as sc
import matplotlib.pyplot as mp
import numpy as np
%matplotlib inline

x= []
with open("./data/multiple3.txt",'r') as f:
    for line in f.readlines():
        data=[float(substr) for substr in line.split(",")]
        x.append(data)      
x=np.array(x)

# 量化带宽
bw = sc.estimate_bandwidth(x,n_samples=len(x),quantile=0.1)

model = sc.MeanShift(bandwidth=bw,bin_seeding=True)
model.fit(x)
centers=model.cluster_centers_


l,r,h = x[:,0].min()-1,x[:,0].max()+1,0.005

b,t,v = x[:,1].min()-1,x[:,1].max()+1,0.005

grid_x=np.meshgrid(np.arange(l,r,h),np.arange(b,t,v))

flat_x=np.c_[grid_x[0].ravel(),grid_x[1].ravel()]
flat_y=model.predict(flat_x)
grid_y=flat_y.reshape(grid_x[0].shape)

pred_y=model.predict(x)

mp.pcolormesh(grid_x[0],grid_x[1],grid_y,cmap='gray')
mp.scatter(x[:,0],x[:,1],c=pred_y,cmap="brg",s=60)
mp.scatter(centers[:,0],centers[:,1],marker="+",s=200,c='red')
mp.show()
```

### 凝聚层次算法

- 首先将每个样本都视作单独的聚类，如果聚类数大于事先给定的期望值，那么就在这些聚类中寻找距离最近的两个聚类凝聚成更大的聚类，以使总聚类数减少，不断重复以上的过程

```python
import sklearn.cluster as sc
import matplotlib.pyplot as mp
import numpy as np
%matplotlib inline

x= []
with open("./data/multiple3.txt",'r') as f:
    for line in f.readlines():
        data=[float(substr) for substr in line.split(",")]
        x.append(data)      
x=np.array(x)

model =  sc.AgglomerativeClustering(n_clusters=4)

pred_y=model.fit_predict(x)
mp.grid(linestyle=":")
mp.scatter(x[:,0],x[:,1],s=60,c=pred_y,cmap='RdBu')
mp.show()
```

### 评价聚类效果的指标

- 好的聚类：内密外疏，同一个聚类中样本足够密集，不同类的足够疏远  
- 轮廓系数  
 a:一个样本与同类其他样本的平均距离。  
 b:一个样本与最近的另一个聚类中各个样本的平均距离。  
 该样本的轮廓系数 s= (b-a)/max(a,b) 
 
- 整个模型的轮廓平均数 = 所有样本的轮廓系数的平均值  
 好的聚类， s 接近于1，  
 差的聚类， s 接近于-1，  
 |s| --> 0 没有聚类的意义，聚类重叠

```python
import sklearn.cluster as sc
import sklearn.metrics as sm
import matplotlib.pyplot as mp
import numpy as np
%matplotlib inline

x= []

with open("./data/multiple3.txt",'r') as f:
    for line in f.readlines():
        data=[float(substr) for substr in line.split(",")]
        x.append(data)      
x=np.array(x)

classtrs,scores,models= np.arange(2,11),[],[]

for n_clusters in classtrs:
    
    model = sc.KMeans(init="k-means++",n_clusters=n_clusters)
    model.fit(x)
    score = sm.silhouette_score(x,model.labels_,sample_size=len(x),metric="euclidean")
    scores.append(score)
    models.append(model)
scores=np.array(scores)   
# print(scores)
best_index=scores.argmax()
best_classtr=classtrs[best_index]
# centers=model.cluster_centers_
best_model=models[best_index]

# print(best_classtr)

l,r,h = x[:,0].min()-1,x[:,0].max()+1,0.005

b,t,v = x[:,1].min()-1,x[:,1].max()+1,0.005

grid_x=np.meshgrid(np.arange(l,r,h),np.arange(b,t,v))

flat_x=np.c_[grid_x[0].ravel(),grid_x[1].ravel()]
flat_y=best_model.predict(flat_x)
grid_y=flat_y.reshape(grid_x[0].shape)

pred_y=best_model.predict(x)

mp.pcolormesh(grid_x[0],grid_x[1],grid_y,cmap='gray')
mp.scatter(x[:,0],x[:,1],c=pred_y,cmap="brg",s=60)
mp.scatter(centers[:,0],centers[:,1],marker="+",s=200,c='red')
mp.show()
```

### DBSCAN(带噪声的基于密度的聚类)算法

随机选择一个样本，作为圆心，以事先给定的半径作圆，被该圆圈中的样本即与中心样本属于同一个聚类，再以这些样本为圆心重复以上的步骤不断扩大聚类的范围，直到再无新样本加入为止，至此完成一个聚类的划分，使用同样的方法，在其余未被聚类圈中的样本，选择新的圆心，得到其他聚类

- 外围样本：被划分某个聚类中，但不能为该聚类引入新的样本
- 孤立样本：不属于任何一个聚类的样本

```python
import sklearn.cluster as sc
import sklearn.metrics as sm
import matplotlib.pyplot as mp
import numpy as np
%matplotlib inline

x=[]
with open("./data/perf.txt",'r') as f:
    for line in f.readlines():
        data=[float(substr) for substr in line.split(",")]
        x.append(data)      
x=np.array(x)

epsilons,scores,models = np.linspace(0.3,1.2,10),[],[]

for epsilon in epsilons:
    model = sc.DBSCAN(eps=epsilon,min_samples=5)
    model.fit( x)
    score = sm.silhouette_score(x,model.labels_,sample_size=len(x),metric="euclidean")
    scores.append(score)
    models.append(model)

scores=np.array(scores)   

best_index=scores.argmax()

best_epsilon=epsilons[best_index]

best_model=models[best_index]

pred_y=best_model.fit_predict(x)
# print(pred_y)
core_mask = np.zeros(len(x),dtype=bool)
core_mask[best_model.core_sample_indices_]=True

offset_maks = best_model.labels_ == -1


periphery_mask = ~(core_mask | offset_maks)

mp.title("DBSCAN",fontsize=14)
labels=set(pred_y)
cs = mp.get_cmap("brg",len(labels))(range(len(labels)))

mp.scatter(x[core_mask][:,0],x[core_mask][:,1],c=cs[pred_y[core_mask]],cmap="brg",s=60,label="core")
mp.scatter(x[periphery_mask][:,0],x[periphery_mask][:,1],edgecolors=cs[pred_y[core_mask]],facecolor="none",cmap="brg",s=60,label="per")
mp.scatter(x[offset_maks][:,0],x[offset_maks][:,1],edgecolors=cs[pred_y[core_mask]],marker="+",s=100,label="off",c='green')
mp.legend()
mp.show()
```
